# Монолитное ядро (макроядро)

Все части монолитного ядра работают в одном адресном пространстве. Это такая схема операционной системы, при которой все компоненты ее ядра являются составными частями программы, используют общие структуры данных и взаимодействуют друг с другом путем непосредственного вызова процедур. Монолитное ядро - тарейший способ организации опреционных систем.

Достоинства:
- Скорость работы
- Упрощенная разработка модулей

Недостатки:
- Все ядро работает в одном адресном пространстве, сбой в одном из компонентов может нарушить работоспособность всей системы.

Предпочтительна для использования ОС, которые должны поддерживать динамические функции и приложения.

# Многослойная архитектура ядра
Архитектура операционной системы, при которой ядро операционной системы разделяется на слои, при этом каждый слой зависит только от тех слоев, что находятся ниже. Разделение на слои происходит условно и может различаться в различных реализациях. Обычно самым верхнем слоем операционной системы является слой, предоставляющий интерфейс системных выховов, обеспечивающий взаимодействие пользовательских приложений и ядра ОС. Нижние слои опреационнной системы обеспечивают взаимодействие программных и аппаратных средств. Это обеспечивает переносимость ОС без необходимости перекомпиляции всего ядра.

Достоинства:
- Различные слои ОС связаны интерфейсами, за счет чего не имеют прямой зависимости друг от друга, при модифицировании отдельного слоя остальные слои не будут затронуты.

Недостатки:
- Сложность проектирования
- При большом числе слоев возникают задержки в работе ядра ОС

Данный тип архитектуры ядра подходит для систем, где модульность и расширяемость являюстя вазными требованиями. 

# Микроядерная архитектура

Суть микроядерной архитектуры заключается в следующем: в привилегированном режиме остается работать очень небольшая часть ОС, называемая микроядром. Микроядро защищено от остальных частей ОС и проложений. В состав микроядра обычно входят машинно-зависимые модули, а также модули, выполняющие базовые функции ядра по управлению процессами, обработке прерываний, управлению виртуальной памятью, перевылке сообщений и управлению устройствами ввода-вывода, связанные с загрузкой или чтением регистров.

Приемущества:
- Высокая стпень переносимости
- Высокая начежность
- Расширяемость

Недостатки:
- Низкая производительность

Данный тип архитектуры ядра подходит системам, в первую очередь ориентированным на надежность.
