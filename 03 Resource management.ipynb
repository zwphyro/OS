# Управление ресурсами со стороны операционной системы

ОС выполняет следуещие **операции для управления ресурсами** (*вне зависимости от типа ресурса*):
- Планирование ресурса
- Удовлетворение запросов на ресурс
- Отслеживание и учет состояния ресурса
- Разрешение конфликтов между потребителями ресурсов

**_Замечание._**  
Потребление ресурсов в мультипрограмной системе ности случайный характер.

# Функциональные компоненты локальной ОС
Основное предоставление состава ОС $-$ группирование по функциональному признаку в соответсвии с типами локальных ресурсов:
- Подсистема управления процессами
- Посдистема управления памятью
- Файловая система
- Система ввода вывода
- Пользовательский интерфейс
- Программные сресдства защины данных и администрирования
- Middleware
- Множественные среды
- Системы и среды программирования
- Пратформы и среды разработки
- Виктуальные машины
