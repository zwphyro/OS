# Системный вызов

Взаимодействие прикладных задач с ядром происходит последством стандартного интерфейса *системных вызовов*.

**_Определение._**  
Интерфейс системных вызовов представляет собой набор услуг ядра и определяет формат запросов на услуги.

Процесс запрашивает услугу посредством системного вызова определенной процедуры ядра, внешне похожего на обычный вызов библиотечной процедуры языка. Ядро от имени процесса выполняет запрос и возвращает к процессу необходимые данные.

# Привилгированный режим

Для надежного управления ходом выполнения приложений система должна иметь по отношению к приложениям определенные *привелегии*.  
Обеспечить привелегии операционной системе несозможно без специальных средств аппаратной поддержки. Аппаратура компьютера должна поддерживать как минимум два режима работы -- пользовательский (*user mode*) и привелегированный режим, который также называют режимом ядра (*kernell mode*), или режим супервизора (*supervisor mode*) Подразумевается, что операционная система или некоторые ее части работают в привелегорованном режиме, а приложения -- в пользовательском.

Так как ядро выполняет основные функции ОС, то чаще всего именно ядро становится той частью ОС, которая работает в привелегированном режиме. Иногда свойство работы в привелегированном режиме используют для определения понятия **ядро**.

# Переход в привилегированный режим посредством системного вызова
Повышение устойчивости операционной системы, обеспечиваемое переходом ядра в привелегированный режим, достигается за счет некоторого замедления выполнения системных вызовов. Системный вызов ядра инициирует переключение процессора из пользовательского режима в привилегированный, а при возврате к приложению -- переключение из привилегированного режима в пользовательский. Во всех типах процессоров из-за дополнительной двукратной задержки переключения переход на процедуру со сменой режима выполняется медленнее, чем процедура без смены режима.

# Реализация системного вызова

Схематично механизм образения к функциям ОС, оформленным в виде серверов, выглядит следующим образом.

![](./images/sys-call-implementation.jpg)

Клиент, которым может быть либо прикладная программа, либо другой компонент ОС, запрашивает выполнение некоторой функции у соответствующего сервера, посылая ему сообщение. Непосредственная передача сообщений между приложениями невозможна, так как из адресные пространства изолированы друг от друга. Микроядро, выполняющееся в привилегированном режие, имеет доступ к адресты пространствам каждого из этих приложений и поэтому может работать в качестве посредника. Микроядро, выполняющееся в привилегированном режиме, имеет доступ к адресным пространствам каждого из этих приложений и поэтому может работать в качестве посредника. Микроядро сначала передает сообшение, содержащее имя и параметры вызываемой процедуры нужному серверу, затем сервер выполняет запрошенную операцию, после чего ядро возвращает результаты клиенту с помощью другого сообщения. Таким образом, работа микроядерной операционной системы соответствует известной модели клиент-сервер. в которой роль транспортных средств выполняет микраядро.
