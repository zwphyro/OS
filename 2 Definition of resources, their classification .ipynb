# Классификация ресурсов

Классификация основных ресурсов операционной системы:
- Аппаратные
- Программные
- Информационные

# Основные аппаратные ресурсы

## Процессорное время
**_Определение._**  
Процессорное время $-$ процесс или набор процессов.

## Оперативная память
Способы разделения оперативной памяти:  
- Временной
- Пространственный

Способы выделения оперативной памяти:
- Статический  
Для фиксированного количества потребителей (вычислительных процессов) априори резервируются области памяти под программные модули и данные.
- Динамический  
Выделение памяти по запросу от процесса как правило, в результате прерывания и в системной области памяти, указание на которую осуществляется через стек.

## Внешняя память
Под определение внешней памяти попадает два ресурса:
- Сама память (ВУ)
- Доступ к памяти (процесс обращения)

Разделение внешней памяти происходит пространственным спсобом.  
Доступ к памяти разделяется временным способом.

## Внешние устройства
Внешние устройства могут быть разделяемыми или неразделяемыми.  
Разделяемые устройства могут разделятся параллельно между различными процессами при помощи механизми прямого доступа.  
Неразделяемые ресурсы работают посредствам последовательного доступа.

# Основные программные ресурсы
**_Определение._**  
Системные программные модули, разделяемые между выполняющимися прцессами/задачами/программами.

## Разделяемые модули
**_Определене._**  
Разделяемые модули $-$ модули, для которых возможно многократное исполнение без искажения кода и данных.

Разделяемые программные ресурсы разделяются на:
- Реентерабельные $-$ модули, которые могут быть прерваны в любой момент с сохранением состояния.
- Повторно-входимые $-$ вход осуществляется по точке входа, выполнение нельзя прервать.



# Информационные ресурсы
К информационным ресурсам возможно получить **доступ по чтению** без специальных средств, или **доступ с возможностью изменения** информационных ресурсов, требующий специальных алгоритмов.
