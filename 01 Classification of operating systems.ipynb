# Типы ОС

**Классификация по различным типам.**  
Операционные системы можно разделить на различные типыx:
- По областям применения
- По методам реализации
- По аппаратным платформам

Также ОС можно разделить на **локальные** и **сетевые**.

**_Определение._**  
ОС компьютера $-$ комплекс взаимосвязанных **управляющих** и **обрабатывающих** программ, который действует как **интерфейс** между приложениями/пользователем с одной стороны и аппаратурой компьютера с другой.

ОС выполняет две основные группы функции:  
- Предоставление расширенной виртуальной машины вместо реальной
- Повышение эффективности работы компьютера путем рационального управления его ресурсами сообразно выбранным критериям

# Классификации ОС
Для операционных систем отсутствует всеобъемлющая классификация.  
В основном ОС классифицирут по **бинарному признаку** сообразно отдельному признаку или применению.

В соответствии с **алгоритмами управления** процессом:
- Многозадачные/однозадачные  
    - С разделением времени
    - Реального времени
    - С вытесняющей и невытесняющей многозадачностью
- Многопоточные/однопоточные
- Многопроцессорные/однопроцессорные


# Другие способы классификации

По назначению:
- Универсальные
- Специализированные

По способу загрузки:
- Загружаемые (выполняют функции после счинывания с некоторого накопителя)
- Резидентные (постоянно загружены в память)

В соответствии с особенностями управления ресурсами:
- По управлению процессорным временем
    - Однозадачные
    - Многозадачные
- По видам многозадачности (способ разделения процессорного времени)
    - Не вытесняющие (процесс выполняется пока сам не освободит ресурс)
    - Вытесняющие (ОС распределяет кванты времени для всех задач)
    
Поддержка работы пользователя:
- Однопользовательские
- Многомользовательские
